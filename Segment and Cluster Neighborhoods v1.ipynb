```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
```

###  Assume the target table is the 1st table in that web page

```python
table = soup.find('table')
rows = table.find_all('tr')
head = table.find_all('th')
```

```python
col = []
col_name = [ele.text.strip() for ele in head]
for ele in col_name:
    col.append((ele,[]))
```

```python
for i in range(1,len(rows)):
    T = rows[i]
    c = 0
    for j in T.find_all('td'):
        ctnt = j.text.strip()
        col[c][1].append(ctnt)
        c +=1
```

```python
Dict = {col:data for (col,data) in col}
df = pd.DataFrame(Dict)
df.head(10)
```

```python
df = df[df['Borough']!='Not assigned']
df.head(10)
```

```python
df.loc[df['Neighbourhood']=='Not assigned','Neighbourhood'] = df['Borough']
df.head(10)
```

```python
df2 = df.groupby(['Postcode','Borough'],sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2.head(10)
```
